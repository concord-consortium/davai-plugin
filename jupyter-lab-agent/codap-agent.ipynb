# An Agent that can control a CODAP document

```python
from strands import Agent
import codap_tool
from codap_tool import codap_request_via_ably

with open("instructions.md", "r") as f1:
    instructions = f1.read().strip()

with open("codap_api.md", "r") as f2:
    codap_api = f2.read().strip()

agent = Agent(
    system_prompt=f"{instructions}\n\n{codap_api}",
    model="us.anthropic.claude-3-7-sonnet-20250219-v1:0",
    tools=[codap_request_via_ably],
)

response = agent(
    "Ask the CODAP plugin to list datasets and summarize them. If there are not datasets create a sample one in the document.",
)
```
